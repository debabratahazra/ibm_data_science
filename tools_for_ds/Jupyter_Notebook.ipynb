### Commonly Used Data Science Libraries

Some of the most commonly used libraries in data science include:

*   **NumPy:** Essential for numerical computing in Python, providing support for large, multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these arrays.
*   **Pandas:** A powerful library for data manipulation and analysis, offering data structures like DataFrames that are highly efficient for handling structured data.
*   **Matplotlib:** A comprehensive library for creating static, animated, and interactive visualizations in Python.
*   **Seaborn:** Built on top of Matplotlib, Seaborn provides a high-level interface for drawing attractive and informative statistical graphics.
*   **Scikit-learn:** A robust and widely used machine learning library offering various classification, regression, clustering, and dimensionality reduction algorithms.
*   **TensorFlow:** An open-source machine learning framework developed by Google, primarily used for deep learning and neural networks.
*   **PyTorch:** Another popular open-source machine learning library, known for its flexibility and dynamic computational graph, often favored for research and rapid prototyping in deep learning.
*   **SciPy:** Provides a collection of algorithms and tools for scientific and technical computing, building on NumPy.

```python
def calculate_result(num1, num2, num3):
  """
  Multiplies num1 and num2, then adds num3 to the product.

  Args:
    num1: The first number for multiplication.
    num2: The second number for multiplication.
    num3: The number to be added.

  Returns:
    The result of (num1 * num2) + num3.
  """
  product = num1 * num2
  result = product + num3
  return result

# Example usage:
number1 = 5
number2 = 10
number3 = 3

final_result = calculate_result(number1, number2, number3)
print(f"The result of ({number1} * {number2}) + {number3} is: {final_result}")

# Another example:
another_result = calculate_result(2, 7, 1)
print(f"The result of (2 * 7) + 1 is: {another_result}")
```

```python
def convert_minutes_to_hours(minutes):
  """
  Converts a given number of minutes into hours.

  Args:
    minutes: The number of minutes to convert (integer or float).

  Returns:
    The equivalent number of hours (float).
  """
  hours = minutes / 60
  return hours

# Example usage:
total_minutes = 150
hours_result = convert_minutes_to_hours(total_minutes)
print(f"{total_minutes} minutes is equal to {hours_result} hours.")

total_minutes_2 = 45
hours_result_2 = convert_minutes_to_hours(total_minutes_2)
print(f"{total_minutes_2} minutes is equal to {hours_result_2} hours.")
```
