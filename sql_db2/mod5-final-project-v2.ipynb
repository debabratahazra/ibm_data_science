### Store the datasets in database tables

To analyze the data using SQL, it first needs to be loaded into SQLite DB.
We will create three tables in as under:

1.  **CENSUS_DATA**
2.  **CHICAGO_PUBLIC_SCHOOLS**
3.  **CHICAGO_CRIME_DATA**


Load the `pandas` and `sqlite3` libraries and establish a connection to `FinalDB.db`


```python
# Import libraries and connect to SQLite DB
import os
import pandas as pd
import sqlite3

workspace_dir = r"c:\gitrepo\GitHub\ibm_data_science\sql_db2"
db_path = os.path.join(workspace_dir, "FinalDB.db")
conn = sqlite3.connect(db_path)
conn
```

Load the SQL magic module


```python
# Load SQL magic
from IPython import get_ipython
get_ipython().run_line_magic('load_ext', 'sql')
```

Use `Pandas` to load the data available in the links above to dataframes. Use these dataframes to load data on to the database `FinalDB.db` as required tables.


```python
# Load local CSVs and write to SQLite tables
csv_dir = workspace_dir

census_csv = os.path.join(csv_dir, "ChicagoCensusData.csv")
schools_csv = os.path.join(csv_dir, "ChicagoPublicSchools.csv")
crime_csv = os.path.join(csv_dir, "ChicagoCrimeData.csv")

df_census = pd.read_csv(census_csv)
df_schools = pd.read_csv(schools_csv)
df_crime = pd.read_csv(crime_csv)

# Normalize school column names to be SQL-friendly
df_schools.columns = [c.strip().replace(' ', '_') for c in df_schools.columns]
if 'Safety_Score' in df_schools.columns:
    df_schools['Safety_Score'] = pd.to_numeric(df_schools['Safety_Score'], errors='coerce')

# Write to SQLite tables (replace if they exist)
df_census.to_sql("CENSUS_DATA", conn, if_exists="replace", index=False)
df_schools.to_sql("CHICAGO_PUBLIC_SCHOOLS", conn, if_exists="replace", index=False)
df_crime.to_sql("CHICAGO_CRIME_DATA", conn, if_exists="replace", index=False)

# Quick sanity prints (first 3 rows)
print("CENSUS_DATA:")
display(df_census.head(3))
print("CHICAGO_PUBLIC_SCHOOLS:")
display(df_schools.head(3))
print("CHICAGO_CRIME_DATA:")
display(df_crime.head(3))
```

Establish a connection between SQL magic module and the database `FinalDB.db`


```python
# Bind SQL magic to the SQLite database
from IPython import get_ipython
get_ipython().run_line_magic('sql', f"sqlite:///{db_path}")
```

Now write and execute SQL queries to solve assignment problems

### Problem 1

##### Find the total number of crimes recorded in the CRIME table.


```python
# Problem 1: Total number of crimes
from IPython import get_ipython
get_ipython().run_line_magic('sql', "SELECT COUNT(*) AS total_crimes FROM CHICAGO_CRIME_DATA;")
```

### Problem 2

##### List community area names and numbers with per capita income less than 11000.


```python
# Problem 2: Community areas with per capita income < 11000
from IPython import get_ipython
get_ipython().run_line_magic('sql', """
SELECT community_area_name, community_area_number
FROM CENSUS_DATA
WHERE per_capita_income < 11000
ORDER BY per_capita_income ASC;
""")
```

### Problem 3

##### List all case numbers for crimes involving minors?(children are not considered minors for the purposes of crime analysis) 


```python
# Problem 3: Case numbers for crimes involving minors (exclude children)
from IPython import get_ipython
get_ipython().run_line_magic('sql', """
SELECT DISTINCT case_number
FROM CHICAGO_CRIME_DATA
WHERE description LIKE '%MINOR%';
""")
```

### Problem 4

##### List all kidnapping crimes involving a child?


```python
# Problem 4: Kidnapping crimes involving a child
from IPython import get_ipython
get_ipython().run_line_magic('sql', """
SELECT case_number, primary_type, description
FROM CHICAGO_CRIME_DATA
WHERE primary_type = 'KIDNAPPING' AND description LIKE '%CHILD%';
"""
)
```

### Problem 5

##### List the kind of crimes that were recorded at schools. (No repetitions)


```python
# Problem 5: Kinds of crimes recorded at schools (no repetitions)
from IPython import get_ipython
get_ipython().run_line_magic('sql', """
SELECT DISTINCT primary_type
FROM CHICAGO_CRIME_DATA
WHERE location_description LIKE '%SCHOOL%'
ORDER BY primary_type;
""")
```

### Problem 6

##### List the type of schools along with the average safety score for each type.


```python
# Problem 6: Type of schools with average safety score (using actual columns)
from IPython import get_ipython
get_ipython().run_line_magic('sql', """
SELECT `Elementary,_Middle,_or_High_School` AS School_Type,
       ROUND(AVG(CAST(SAFETY_SCORE AS REAL)), 2) AS avg_safety_score
FROM CHICAGO_PUBLIC_SCHOOLS
GROUP BY `Elementary,_Middle,_or_High_School`
ORDER BY avg_safety_score DESC;
""")
```

### Problem 7

##### List 5 community areas with highest % of households below poverty line


```python
# Problem 7: Top 5 community areas by % households below poverty line
from IPython import get_ipython
get_ipython().run_line_magic('sql', """
SELECT community_area_name, community_area_number, percent_households_below_poverty
FROM CENSUS_DATA
ORDER BY percent_households_below_poverty DESC
LIMIT 5;
""")
```

### Problem 8

##### Which community area is most crime prone? Display the coumminty area number only.


```python
# Problem 8: Most crime-prone community area number
from IPython import get_ipython
get_ipython().run_line_magic('sql', """
SELECT community_area_number
FROM CHICAGO_CRIME_DATA
GROUP BY community_area_number
ORDER BY COUNT(*) DESC
LIMIT 1;
""")
```

### Problem 9

##### Use a sub-query to find the name of the community area with highest hardship index


```python
# Problem 9: Community area name with highest hardship index (subquery)
from IPython import get_ipython
get_ipython().run_line_magic('sql', """
SELECT community_area_name
FROM CENSUS_DATA
WHERE hardship_index = (
    SELECT MAX(hardship_index) FROM CENSUS_DATA
);
""")
```

### Problem 10

##### Use a sub-query to determine the Community Area Name with most number of crimes?


```python
# Problem 10: Community Area Name with most number of crimes (subquery)
from IPython import get_ipython
get_ipython().run_line_magic('sql', """
SELECT c.community_area_name
FROM CENSUS_DATA c
WHERE c.community_area_number = (
    SELECT community_area_number
    FROM CHICAGO_CRIME_DATA
    GROUP BY community_area_number
    ORDER BY COUNT(*) DESC
    LIMIT 1
);
""")
```
